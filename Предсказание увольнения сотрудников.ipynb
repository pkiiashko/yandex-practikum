```python
def plot_histograms_boxplots(df):
    """
    –°—Ç—Ä–æ–∏—Ç –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ —è—â–∏–∫–∏ —Å —É—Å–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∞ –≤ DataFrame.

    :param df: DataFrame, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –¥–∞–Ω–Ω—ã–µ
    """
    # –í—ã–±–æ—Ä —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    numeric_cols = df.select_dtypes(include=['number']).columns

    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ —Ñ–∏–≥—É—Ä—ã
    num_plots = len(numeric_cols)
    plt.figure(figsize=(10, 5 * num_plots))

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º –∏ —è—â–∏–∫–æ–≤ —Å —É—Å–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
    for i, col in enumerate(numeric_cols):
        # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
        plt.subplot(num_plots, 2, 2*i + 1)
        df[col].hist(bins=15)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')

        # –Ø—â–∏–∫ —Å —É—Å–∞–º–∏
        plt.subplot(num_plots, 2, 2*i + 2)
        df.boxplot(column=col)
        plt.title(f'Box Plot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
def plot_countplots(df):
    """
    –°—Ç—Ä–æ–∏—Ç countplots –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∞ –≤ DataFrame.

    :param df: DataFrame, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –¥–∞–Ω–Ω—ã–µ.
    """
    # –í—ã–±–æ—Ä –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    categorical_cols = df.select_dtypes(exclude=['number']).columns

    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ —Ñ–∏–≥—É—Ä—ã
    num_plots = len(categorical_cols)
    plt.figure(figsize=(10, 5 * num_plots))

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ countplots –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
    for i, col in enumerate(categorical_cols):
        plt.subplot(num_plots, 1, i + 1)
        sns.countplot(y=col, data=df, order=df[col].value_counts(ascending=False).index)
        plt.title(f'–ì—Ä–∞—Ñ–∏–∫ {col}')
        plt.xlabel(col)
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

    plt.tight_layout()
    plt.show()
```

## –®–∞–≥ –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–û–±—É—á–∏—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–µ –º–æ–¥–µ–ª–∏. –í–æ–∑—å–º–∏—Ç–µ –æ–¥–Ω—É –ª–∏–Ω–µ–π–Ω—É—é –º–æ–¥–µ–ª—å, –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –≤—Ç–æ—Ä–æ–π ‚Äî –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π. –ü–æ–¥–±–µ—Ä–∏—Ç–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –æ–¥–Ω–æ–π –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é –æ–¥–Ω–æ–≥–æ –∏–∑ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –≤–∞–º –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤.
–í—ã–±–µ—Ä–∏—Ç–µ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ. –í—ã–±–æ—Ä –¥–µ–ª–∞–π—Ç–µ –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ ‚Äî SMAPE (–∞–Ω–≥–ª. symmetric mean absolute percentage error, ¬´—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ¬ª).

```python
X_train = train_job_rate.drop(['job_satisfaction_rate'], axis=1)
y_train = train_job_rate['job_satisfaction_rate']
X_test = test_features
y_test = test_target_rate

ohe_columns = X_train.select_dtypes(exclude=['number']).columns
num_columns = X_train.select_dtypes(include=['number']).columns
ord_columns = ['last_year_promo','last_year_violations']

ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse=False))
    ]
    )
# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# SimpleImputer + OE
ord_pipe = Pipeline([
    (
        'ord',
        OrdinalEncoder(
            handle_unknown='use_encoded_value',
            unknown_value=np.nan
        )
    ),
    (
        'simple_Imputer',
        SimpleImputer(
            missing_values=np.nan, 
            strategy='most_frequent'
        )
    )
])


data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])

param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 1000),
        'models__max_features': range(0, 220), 
        'models__min_samples_leaf': range(2, 15),
        'models__min_samples_split': range(2, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LinearRegression() 
    {
        'models': [LinearRegression()],
        'models__n_features': range(0, 1000), 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'] 
    },
]
pipe_final.fit(X_train, y_train)

def smape(y_test, y_pred):
    numerator = np.abs(y_pred - y_test)
    denominator = (np.abs(y_test)+np.abs(y_pred))/2
    smape_value = np.mean(numerator/denominator)*100
    return smape_value

smape_s = make_scorer(smape,   greater_is_better=False)

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    scoring=smape_s,
    cv=5,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter=100
)
    
randomized_search.fit(X_train, y_train)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search.best_score_)
```

```python
preds_final = randomized_search.predict(X_test)
```

```python
preds = randomized_search.predict(X_train)
```

```python
smape_best = smape(preds, y_train.to_numpy())
print('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', smape_best.round(2))
```

```python
def plot_compare(df):

# –í—ã–±–æ—Ä —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    numeric_cols = df.select_dtypes(include=['number']).columns

    num_plots = len(numeric_cols)
    plt.figure(figsize=(10, 5 * num_plots))

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
    for i, col in enumerate(numeric_cols):
        # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
        plt.subplot(num_plots, 2, 2*i + 1)
        df.loc[df['quit']=='yes'][col].hist(bins=15,alpha=0.5,label='—É–≤–æ–ª–∏–ª–∏—Å—å')
        df.loc[df['quit']=='no'][col].hist(bins=15,alpha=0.5,label='–Ω–µ —É–≤–æ–ª–∏–ª–∏—Å—å')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.legend(loc='upper right')
plt.show()        
```

```python
def plot_pivot(df):
    """
    –°—Ç—Ä–æ–∏—Ç countplots –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∞ –≤ DataFrame.

    :param df: DataFrame, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –¥–∞–Ω–Ω—ã–µ.
    """
    # –í—ã–±–æ—Ä –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
    categorical_cols = df.select_dtypes(exclude=['number']).columns
    categorical_cols = categorical_cols.drop(['quit'])

    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ —Ñ–∏–≥—É—Ä—ã
    num_plots = len(categorical_cols)
    plt.figure(figsize=(10, 5 * num_plots))

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ countplots –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
    for i, col in enumerate(categorical_cols):
        plt.subplot(num_plots, 1, i + 1)
        pivot_1 = pd.pivot_table(train_quit,
               index=col, values=["salary"], aggfunc=len)
        pivot_2 = pd.pivot_table(train_quit.loc[df['quit']=='yes'],
               index=col, values=["salary"], aggfunc=len)
        pivot_3 = pivot_2/pivot_1
        pivot_3 = pivot_3.reset_index()  
        sns.barplot(y=col,x= pivot_3['salary'], data=pivot_3, order=df[col].value_counts(ascending=False).index)
        plt.title(f'–ì—Ä–∞—Ñ–∏–∫ {col}')
        plt.xlabel(col)
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
        display(pivot_3)

    plt.tight_layout()
    plt.show()
```

```python
plot_pivot(train_quit)
```

–ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ç–∞–∫ –∂–µ, –∫–∞–∫ –∏ –≤ –ø–µ—Ä–≤–æ–π –∑–∞–¥–∞—á–µ: –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –≤ –ø–∞–π–ø–ª–∞–π–Ω–µ, –¥–æ–ø–æ–ª–Ω–∏–≤ –ø–∞–π–ø–ª–∞–π–Ω –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏. –ü—Ä–∏ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–∏ —É—á–∏—Ç—ã–≤–∞–π—Ç–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –º–æ–¥–µ–ª–µ–π –∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–∞ –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞.

–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ –Ω–∏–∂–µ –≤ –ø–∞–π–ø–ª–∞–π–Ω–µ

## –®–∞–≥ –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

–û–±—É—á–∏—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º —Ç—Ä–∏ –º–æ–¥–µ–ª–∏. –ö–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –¥–≤—É—Ö –∏–∑ –Ω–∏—Ö –ø–æ–¥–±–µ—Ä–∏—Ç–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∫–∞—á–µ—Å—Ç–≤–æ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏.
–ú–µ—Ç—Ä–∏–∫–∞ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –≤ —ç—Ç–æ–π –∑–∞–¥–∞—á–µ ‚Äî ROC-AUC. –ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å–ø–µ—Ö–∞: ROC-AUC ‚â•0.91 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. –ù–∞–ø–æ–º–Ω–∏–º: –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —á–∞—Å—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç —É–ª—É—á—à–∏—Ç—å –º–µ—Ç—Ä–∏–∫—É.

```python
X_train = train_quit.drop(['quit'], axis=1)
y_train = train_quit['quit']
X_test = test_features_all_quit.drop(['quit'], axis=1)
y_test = test_features_all_quit['quit']

ohe_columns = X_train.select_dtypes(exclude=['number']).columns
num_columns = X_train.select_dtypes(include=['number']).columns
ord_columns = ['last_year_violations']

ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse=False))
    ]
    )

# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤
# SimpleImputer + OE
# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# SimpleImputer + OE
ord_pipe = Pipeline([
    (
        'ord',
        OrdinalEncoder(
            handle_unknown='use_encoded_value',
            unknown_value=np.nan
        )
    ),
    (
        'simple_Imputer',
        SimpleImputer(
            missing_values=np.nan, 
            strategy='most_frequent'
        )
    )
])


data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 10),
        'models__max_features': range(2, 10), 
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, probability= True)],
        'models__n_features': range(1, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
pipe_final.fit(X_train, y_train) 
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search.best_score_)
preds_final = randomized_search.predict_proba(X_test)[:,1]
```

```python
roc_auc_best = roc_auc_score(y_test,preds_final)
print('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:', roc_auc_best.round(2))
```
